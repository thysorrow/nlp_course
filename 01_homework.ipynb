### Задание 4

Извлеките как можно больше (минимум 30) глаголов из текста одним регулярным выражением, не используя основы в самом выражении.

Регулярное выражение должно быть написано на нескольких строчках с комментариями (используйте флаг re.VERBOSE )

```python
# нужно описать все окончания русских глаголов. при таком паттерне будут также находиться неглаголы у которых такие же окончания

pattern = re.compile(r"""
(\b\w*а(ю|ем)\b)           |   # настоящее и будущее времена, 1 лицо
(\b\w*(ешь|ишь|ете|ите)\b) |   # настоящее и будущее времена, 2 лицо
(\b\w*(ет|ют|ёшь|ут)\b)    |   # настоящее и будущее времена, 3 лицо
(\b\w*л[аои]?\b)           |   # прошедшее время
(\b\w*(ся|сь)\b)               # возвратные глаголы всех времён
""",re.VERBOSE)

verbs = re.findall(pattern,text)
len(verbs)
verbs
```
